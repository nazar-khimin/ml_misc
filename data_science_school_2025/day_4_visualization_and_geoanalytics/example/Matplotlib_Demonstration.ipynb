## Демонстрація роботи з бібліотекою Matplotlib
### 1. Основи використання Matplotlib

**Matplotlib** — це бібліотека для Python, яка використовується для створення 2D графіків і візуалізацій даних. Її головна перевага — простота використання та велика кількість можливостей для налаштування графіків. Вона дозволяє створювати різні типи графіків, такі як лінійні, гістограми, кругові діаграми тощо, що допомагає візуалізувати дані для досліджень або презентацій.

#### 1.1 Проста лінійна діаграма

```python
import matplotlib.pyplot as plt

# Дані
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Створення графіка
plt.plot(x, y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.show()
```

#### 1.2 Гістограма

```python
import numpy as np
# Генеруємо випадкові дані
data = np.random.normal(0, 1, 1000)
# data = np.random.random(1000)

# Створюємо гістограму
plt.hist(data, bins='auto', alpha=0.7, color='blue')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

#### 1.3 Кругова діаграма

```python
# Дані
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]

# Створення кругової діаграми
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

### 2. Додаткові властивості Matplotlib

#### 2.1 Зміна кольорів, маркерів і стилів ліній

```python
# Встановимо стиль 'seaborn'
plt.style.use('seaborn')

# Створимо той же графік з новим стилем
plt.plot(x, y, marker='o', color='green', linestyle='--')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Styled Line Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Дані
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 3, 5, 7, 11]

# Побудова графіків з різними стилями
plt.plot(x, y1, color='blue', linestyle='-', linewidth=2, marker='o', markersize=8, label='y = x^2')
plt.plot(x, y2, color='red', linestyle='--', linewidth=2, marker='s', markersize=8, label='y = prime')

# Додавання підписів та заголовків
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Customized Line Plot')
plt.legend()  # Показує легенду
plt.show()
```

#### 2.2 Додавання сітки та обмеження осей

```python
# Додавання сітки та обмеження для осей
plt.plot(x, y1, label='y = x^2')
plt.plot(x, y2, label='y = prime')

# Встановлюємо обмеження для осей
plt.xlim(1, 5)
plt.ylim(0, 30)

# Додаємо сітку
plt.grid(True)

plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot with Grid and Axis Limits')
plt.legend()
plt.show()
```

#### 2.3 Субплоти (Subplots)

```python
# Створимо фігуру з двома субплотами
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Перший субплот
ax1.plot(x, y1, color='purple', marker='o')
ax1.set_title('y = x^2')

# Другий субплот
ax2.plot(x, y2, color='green', marker='s')
ax2.set_title('y = prime')

# Додаємо підписи
ax1.set_xlabel('X Axis')
ax1.set_ylabel('Y Axis')
ax2.set_xlabel('X Axis')
ax2.set_ylabel('Y Axis')

plt.suptitle('Multiple Subplots')
plt.show()
```

#### 2.4 Анотації на графіку

```python
plt.plot(x, y1, label='y = x^2')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot with Annotations')

# Додаємо анотацію для певної точки
plt.annotate('Special Point', xy=(3, 9), xytext=(3, 15),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.legend()
plt.show()
```

#### 2.5 Теплова карта (Colormaps)

```python
import numpy as np

# Генеруємо випадкові дані для теплової карти
data = np.random.rand(10, 10)

# Відображення теплової карти з колірною картою 'viridis'
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar()  # Додає шкалу кольорів
plt.title('Heatmap with Colormap')
plt.show()
```

#### 2.6 3D Графіки

```python
from mpl_toolkits.mplot3d import Axes3D

# Дані для 3D графіка
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Створення 3D фігури
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Побудова поверхні
ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_title('3D Surface Plot')
plt.show()
```

#### Налаштування осей та подвійна шкала Y

```python
import matplotlib.pyplot as plt
import numpy as np

# Дані для графіка
years = np.arange(2010, 2021)
population = [2.5, 2.7, 2.9, 3.0, 3.2, 3.3, 3.5, 3.6, 3.7, 3.8, 3.9]  # в мільйонах
unemployment_rate = [78, 74, 69, 67, 63, 60, 57, 55, 53, 51, 48]  # у відсотках

# Створення фігури і основної осі
fig, ax1 = plt.subplots(figsize=(10, 6))

# Налаштування першої осі Y (лівої) для показу населення
ax1.plot(years, population, color='blue', marker='o', label="Population (millions)")
ax1.set_xlabel("Year")  # Підпис осі X
ax1.set_ylabel("Population (millions)", color='blue')  # Підпис лівої осі Y
ax1.tick_params(axis='y', labelcolor='blue')  # Колір підписів на лівій осі Y

# Налаштування другої осі Y (правої) для рівня безробіття
ax2 = ax1.twinx()  # Створення другої осі Y, яка поділяє вісь X із першою
ax2.plot(years, unemployment_rate, color='red', marker='s', linestyle='--', label="Unemployment Rate (%)")
ax2.set_ylabel("Unemployment Rate (%)", color='red')  # Підпис правої осі Y
ax2.tick_params(axis='y', labelcolor='red')  # Колір підписів на правій осі Y

# Додаємо заголовок і налаштовуємо сітку
plt.title("Population Growth and Unemployment Rate (2010-2020)")
ax1.grid(True, which='both', axis='both', linestyle=':', linewidth=0.5)

# Додаємо легенду для кожної осі
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```

#### Інтерактивні графіки з Plotly

```python
import plotly.graph_objects as go
import numpy as np

# Створюємо дані
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Створюємо інтерактивний графік
fig = go.Figure()

# Додаємо лінію
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='y = sin(x)'))

# Додаємо заголовок і підписи осей
fig.update_layout(
    title="Interactive Sine Wave",
    xaxis_title="X",
    yaxis_title="Y"
)

# Відображення графіка
fig.show()
```

#### Створення двох осей Y

```python
import plotly.graph_objects as go

# Дані для графіка
x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 17, 20]  # Перша серія даних
y2 = [30, 35, 30, 40, 50]  # Друга серія даних

fig = go.Figure()

# Перша лінія (ліва вісь Y)
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='Line 1', yaxis="y1"))

# Друга лінія (права вісь Y)
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='Line 2', yaxis="y2"))

# Налаштування осей
fig.update_layout(
    title="Plot with Two Y Axes",
    xaxis_title="X Axis",
    yaxis=dict(title="Left Y Axis"),
    yaxis2=dict(title="Right Y Axis", overlaying='y', side='right')
)
fig.show()
```

#### 3D поверхня

```python
import plotly.graph_objects as go
import numpy as np

# Дані для 3D поверхні
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Створюємо 3D поверхню
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale="Viridis")])
fig.update_layout(title="3D Surface Plot", scene=dict(
                    xaxis_title='X Axis',
                    yaxis_title='Y Axis',
                    zaxis_title='Z Axis'))
fig.show()
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objs as go
import numpy as np

# Ініціалізація додатка Dash
app = Dash(__name__)

# Оформлення дашборду
app.layout = html.Div([
    html.H1("Dash Example: Dynamic Sine Wave"),
    
    # Повзунок для вибору амплітуди
    html.Label("Select Amplitude:"),
    dcc.Slider(
        id='amplitude-slider',
        min=0.5,
        max=5,
        step=0.1,
        value=1,  # Початкове значення
        marks={i: f'{i}' for i in range(1, 6)}
    ),
    
    # Динамічний графік
    dcc.Graph(id='sine-wave-plot')
])

# Оновлення графіка при зміні значення повзунка
@app.callback(
    Output('sine-wave-plot', 'figure'),
    [Input('amplitude-slider', 'value')]
)
def update_graph(amplitude):
    # Генерація даних для синусоїдної хвилі
    x = np.linspace(0, 10, 100)
    y = amplitude * np.sin(x)
    
    # Оновлення графіка
    fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', line=dict(color='blue')))
    fig.update_layout(title=f"Sine Wave with Amplitude {amplitude}",
                      xaxis_title="X Axis",
                      yaxis_title="Y Axis")
    return fig

# Запуск дашборду
# дашборд буде доступний за адресою http://127.0.0.1:8050/ у веб-браузері.
if __name__ == '__main__':
    app.run_server(debug=True)
```

#### Просунутий приклад дашборда

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Зразок даних
df = px.data.gapminder()

# Ініціалізація додатка
app = dash.Dash(__name__)

# Шаблон додатка
app.layout = html.Div([
    html.H1("Просунутий дашборд"),
    html.Label("Виберіть континент:"),
    dcc.Dropdown(
        id='continent-dropdown',
        options=[{'label': i, 'value': i} for i in df['continent'].unique()],
        value='Asia'
    ),
    html.Label("Виберіть рік:"),
    dcc.Slider(
        id='year-slider',
        min=df['year'].min(),
        max=df['year'].max(),
        value=df['year'].min(),
        marks={str(year): str(year) for year in df['year'].unique()},
        step=None
    ),
    dcc.Graph(id='scatter-graph'),
    dcc.Graph(id='line-graph')
])

# Колбек для оновлення графіків
@app.callback(
    [Output('scatter-graph', 'figure'), Output('line-graph', 'figure')],
    [Input('continent-dropdown', 'value'), Input('year-slider', 'value')]
)
def update_graphs(selected_continent, selected_year):
    # Фільтрація даних за вибраним континентом та роком
    filtered_df = df[(df['continent'] == selected_continent) & (df['year'] == selected_year)]
    
    # Розсіювальний графік залежності ВВП від тривалості життя
    scatter_fig = px.scatter(filtered_df, x='gdpPercap', y='lifeExp', size='pop', color='country',
                             title=f'{selected_continent} у {selected_year}', log_x=True)
    
    # Лінійний графік тривалості життя для континенту з часом
    line_df = df[df['continent'] == selected_continent]
    line_fig = px.line(line_df, x='year', y='lifeExp', color='country',
                       title=f'Зміна тривалості життя у {selected_continent} з часом')
    
    return scatter_fig, line_fig

if __name__ == '__main__':
    app.run_server(debug=True)
```

# Інтеграція дашборду на базі Bokeh з веб-додатком на Django

Для інтеграції дашборду на базі бібліотеки Bokeh з веб-додатком на Django можна скористатися дещо іншим підходом, оскільки Bokeh не має готової інтеграції з Django так, як це зроблено для Plotly. Однак Bokeh надає зручний інтерфейс для вбудовування графіків через HTML-скрипти, які легко інтегруються в шаблони Django.

Ось кроки, як це можна зробити:

## Крок 1: Встановіть Bokeh

Встановіть бібліотеку Bokeh через pip:

<code> pip install bokeh</code>

## Крок 2: Створіть графік Bokeh у контролері (views.py)

У вашому файлі `views.py` створіть графік Bokeh і збережіть його в HTML-компоненті за допомогою функції components:

<code>from django.shortcuts import render
from bokeh.plotting import figure
from bokeh.embed import components</code>

<code>def bokeh_dashboard_view(request):
    # Створення простого графіка
    plot = figure(title="Простий графік Bokeh", x_axis_label="X", y_axis_label="Y")
    plot.line([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], line_width=2)
    # Отримання скрипту та div для вставки в HTML-шаблон
    script, div = components(plot)
    # Передача в шаблон
    return render(request, 'bokeh_dashboard.html', {'script': script, 'div': div})
</code>


## Крок 3: Додайте маршрут до контролера у urls.py

У файлі `urls.py` додайте шлях до вашого контролера:

<code>from django.urls import path
from . import views
urlpatterns = [
    ...
    path('bokeh-dashboard/', views.bokeh_dashboard_view, name='bokeh_dashboard'),
    ...
]
</code>
    
### Крок 4: Створіть HTML-шаблон для відображення графіка Bokeh

У файлі шаблону Django (наприклад, `bokeh_dashboard.html`) додайте наступний код:

`<!DOCTYPE html>`

`<html lang="en">`

`<head>`

`    <meta charset="UTF-8">`

`    <title>Bokeh Dashboard</title>`

`    <link rel="stylesheet" href="https://cdn.bokeh.org/bokeh/release/bokeh-2.4.3.min.css" type="text/css">`

`    <script src="https://cdn.bokeh.org/bokeh/release/bokeh-2.4.3.min.js"></script>`

`</head>`

`<body>`

`    <h1>Інтерактивний дашборд з Bokeh</h1>`

`    <!-- Вставляємо Bokeh-графік -->`

`    {{ script|safe }}`

`    {{ div|safe }}`

`</body>`

`</html>`


### Крок 5: Запустіть сервер

Запустіть сервер Django:

<code>python manage.py runserver</code>

Тепер ваш дашборд на основі Bokeh буде доступний за адресою http://127.0.0.1:8000/bokeh-dashboard/.

Додатково: Динамічне оновлення графіків

Якщо вам потрібен більш інтерактивний досвід із можливістю динамічного оновлення графіків, розгляньте можливість використання WebSocket і Bokeh Server. Однак це значно складніший процес і вимагає додаткової конфігурації.
