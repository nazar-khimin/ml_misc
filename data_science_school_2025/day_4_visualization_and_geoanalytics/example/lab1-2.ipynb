## Line Plots <a id="12"></a>


#### *What is a line plot and why use it?*

***A line chart or line plot*** is a type of plot which displays information as a series of data points called *'markers'* connected by straight line segments. It is a basic type of chart common in many fields.

*Use line plot when you have a continuous data set.* ***These are best suited for trend-based visualizations of data over a period of time.***

---

Before we proceed, our dataframe has more than 16000 rows. It is too huge to visualize the entire time period of the price movement. To deal with it, we should perform ***data resampling***.

#### What is a resampling?
***Resampling*** is a crucial method for time series analysis that enables you to freely choose the desired level of data resolution. You can either upsample, or increase the number of data points, such as by transforming 5-minute data into 1-minute data. 

The basic syntax for resampling in Pandas is `dataframe.resample('desired resolution')` method. Along with that, different aggregation function can be used.

Let's downsample the series from *1-minute* into *1-hour* intervals. Considering the semantics of our dataset: 
*   for column **Open** we take *the first value* of a 1-hour interval
*   for **Close** we have *the last value*. 
*   for **High** *maximum value* within a 1-hour interval is taken
*   for **Low** we take *minimum value* 
*   column **Volume** will store *all summed-up values* within a 1-hour interval
*   for all indicators their mean values within a 1-hour interval will be calculated

````python
# Resample data into 1-hour intervals
df_resampled = df.loc[:, 'Price':'MACD'].resample("1H").agg({
    'Price': 'mean',
    'Volume': 'sum',
    'Open': 'first',
    'High': 'max',
    'Low': 'min',
    'Close': 'last',
    'ADOSC': 'mean',
    'NATR': 'mean', 
    'TRANGE': 'mean', 
    'SMA': 'mean', 
    'MACD': 'mean'
}).dropna()
````

### Line Chart with Multiple Subplots <a id="14"></a>

Cryptocurrency indicators are often plotted on a second subplot. It is done for two main reasons:

*   ***To separate the indicator from the price chart:*** Placing the indicator on a separate subplot allows it to be visually separated from the price chart, making it easier to interpret and analyze.

*   ***To enable comparison:*** By placing the indicator on a separate subplot, it is possible to compare the indicator values with the price values on the main subplot. This can provide valuable insights into price trends and help identify potential buy or sell signals.

Placing the indicator on a separate subplot also allows for flexibility in adjusting the scale of the indicator, which can be different from the scale of the price chart. This is especially important if the indicator values have a significantly different range than the price values, as it can make it easier to compare and analyze both sets of data.

#### Line Subplots Chart with MACD indicator

***MACD*** is a kind of *oscillating indicator.* Oscillator looks into whether the trend is upward or downward. In MACD case, when *the main line is above zero* for a sustained period of time, the asset is likely trending ***upwards***. Conversely, if *the main line is below zero* for a sustained period of time, then the asset is likely under ***downwards.***

***A signal line*** is the second line that is *Moving Average* of the main line of MACD. ***Crossovers with these 2 lines*** (the main line and the signal line) can also provide additional ***buy and sell signals*** in the same calculation of the shorter and the longer Moving Averages crossovers.

##### What are buy and sell signals?
The possible ***buy and sell signals*** are triggered when *the shorter average crosses the longer*. Crossing in the *upward direction* triggers a ***possible buy*** while crossing in the *downward direction* signals a ***possible sell***.

````python
def MACD(df, s, l, signal_period):
    """
    Returns MACD, signal line and their crossovers (buy and sell signals).
    """
    df_resampled.MACD = df_resampled.Close.ewm(span=s, min_periods=1).mean() - df_resampled.Close.ewm(span=l, min_periods=1).mean()
    signal = df_resampled.MACD.ewm(span=signal_period, min_periods=1).mean()
    diff = df_resampled.MACD - signal
    
    return df, signal, diff
````

````python
# Calculate MACD, signal line and their crossovers (buy and sell signals)
df_resampled, signal, diff = MACD(df_resampled, 12, 26, 9)
````

To define several suplots we will use `plt.subplot(nrows, ncols, index)` method, where:

*  `nrows, ncols, index` - the position of the subplot described by one of three integers. The subplot will take the index position on a grid with `nrows` rows and `ncols` columns. `index` starts at 1 in the upper left corner and increases to the right. 

***Buy and sell signals*** we will visualize by a ***scatter plot.***
With ***matplotlib.pyplot***, you can use the `plt.scatter()` function to draw a ***scatter plot.*** The `plt.scatter()` function plots one dot for each observation. It needs two arrays of the same length, one for the values of the *x-axis*, and one for values on the *y-axis*.

````python
# Create a figure and a set of subplots specifying figure size
plt.figure(figsize=(15, 10))
    
# Define the first subplot
# Set title
ax = plt.subplot(2,1,1)
plt.title("Close price")
# Set axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')
# Plot the original Closing price
plt.plot(df_resampled.Close, label="Close")
# Display grid
plt.grid(True)
    
# Define the second subplot
ax = plt.subplot(2,1,2)
# Plot MACD and signal line
plt.plot(df_resampled.MACD, label="MACD")
plt.plot(signal, label="Signal")
    
# Set title of the subplot
plt.title("MACD")
# Set axis labels
ax.set_xlabel('Date')
ax.set_ylabel('MACD value')
# Display legend
plt.legend()
# Display grid
plt.grid(True)

# Crossover points
for i in range(1, len(df_resampled)):
    # Plot buy signals
    if diff[i-1] < 0 and diff[i] > 0:
        plt.scatter(df_resampled.index[i], df_resampled.iloc[i]["MACD"], marker="o", s=100, color="b", alpha=0.9)
    
    # Plot sell signals
    if diff[i-1] > 0 and diff[i] < 0:
        plt.scatter(df_resampled.index[i], df_resampled.iloc[i]["MACD"], marker="o", s=100, color="r", alpha=0.9)
````

### Dual-axis Chart <a id="15"></a>

A graph that shows one curve on the left side of the y-axis and another on the right side of the y-axis is called ***a dual-axis chart.*** This type of chart can provide several benefits, including:

*   ***Showing two related variables on the same chart:*** In some cases, you may want to show two related variables on the same chart, but they have different scales. A dual-axis chart can be useful in this situation because it allows you to show both variables on the same chart, even if they have different units or scales.

*   ***Highlighting relationships between variables:*** A dual-axis chart can help you highlight relationships between two variables that might not be obvious from looking at them individually. By placing them on the same chart, you can see how they change together over time, which can help you identify patterns and trends.

*   ***Improving data visualization:*** A dual-axis chart can help you improve data visualization by reducing the need for multiple charts or tables. Instead of having to look at two separate charts, you can see both variables on the same chart, which can help you better understand the data and make more informed decisions.

Let's visualize *trends movements* through specific time period for ***BTC currency*** and ***ADOSC indicator.*** A dual-axis chart is constructed by plotting two different data series on the same chart, with each series having a different scale or unit of measurement. 

First, we need to set $x$ and $y$ *data point values* for ***Price*** and ***ADOSC*** lines. Along the ***x-axis***, the points will be placed according to ***the index field*** of our dataframe *(according to the datetime).* While the ***y-axis*** will display ***cryptocurrency Price*** and ***ADOSC indicator*** data.

````python
# Set x-axis data
x = df_resampled.index
# Set y1-axis data (left Y-axis)
y1 = df_resampled['Price']
# Set y2-axis data (right Y-axis)
y2 = df_resampled['ADOSC']
````

To adjust the two y-axis we need to use the `axes.twinx()` method that *create a twin axes sharing ***the one x-axis.**** You can access the parameters of each of the axes in the same way by referring to their functions and attributes.

````python
# Create a figure and a set of subplots specifying figure size
fig, ax1 = plt.subplots(figsize=(16,9))

# Plot Line1 (left Y Axis)
ax1.plot(x, y1, color='tab:red')

# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()

# Plot Line2 (right Y Axis)
ax2.plot(x, y2, color='tab:blue')

# Decorations
# ax1 (left Y axis)
# Set x-axis labels
ax1.set_xlabel('Date', fontsize=20)
# Set x-axis labels parameters
ax1.tick_params(axis='x', rotation=0, labelsize=12)
# Set y-axis label
ax1.set_ylabel('Price', color='tab:red', fontsize=20)
# Set y-axis label parameters
ax1.tick_params(axis='y', rotation=0, labelcolor='tab:red' )
ax1.grid(alpha=.4)

# ax2 (right Y axis)
# Set y-axis label
ax2.set_ylabel('ADOSC', color='tab:blue', fontsize=20)
# Set y-axis label parameters
ax2.tick_params(axis='y', rotation=0, labelsize=12, labelcolor='tab:blue')
# Set title
ax2.set_title("Dual-axis Chart for BTC Price and ADOSC indicator", fontsize=22)

# Adjust the padding between and around subplots
fig.tight_layout()

# Show plot
plt.show()
````

### Horizontal lines<a id="16"></a>

#### Support and Resistance Levels

***Support and resistance levels*** are important levels recognizable on a chart, where supply and demand meet. Learning to recognize these levels can help the trader with successful entries and exits. Read more about it [here](https://www.investopedia.com/trading/support-and-resistance-basics/).

We will use fractal candlestick pattern to detect the pivot points which then become the support or resistance levels. ***The bullish fractal*** pattern signals the price could move higher. ***A bearish fractal*** signals the price could move lower. 

A fractal is a candlestick pattern made by 5 candles. The third candle has the lowest low price, the previous candles have decreasing lows and the next candles have increasing lows. *By this pattern, the low of the third candle is the ***support level.**** The same concept applies to ***resistance levels***, where the third candle has *the highest high* of the five ones.


````python
# Determine bullish fractal 
def is_support(df,i):  
    condition1 = df['Low'][i] < df['Low'][i-1] and df['Low'][i] < df['Low'][i+1]      
    condition2 = df['Low'][i+1] < df['Low'][i+2] and df['Low'][i-1] < df['Low'][i-2] 
    return (condition1 and condition2) 

# Determine bearish fractal
def is_resistance(df,i):  
    condition1 = df['High'][i] > df['High'][i-1] and df['High'][i] > df['High'][i+1]      
    condition2 = df['High'][i+1] > df['High'][i+2] and df['High'][i-1] > df['High'][i-2]    
    return (condition1 and condition2)

# Make sure the new level area does not exist already
def is_far_from_level(value, levels, df):    
    average =  np.mean(df['High'] - df['Low'])    
    return np.sum([abs(value-level) < average for _, level in levels]) == 0

def support_resistence(df):
    # A list to store resistance and support levels
    levels = []
    for i in range(2, df.shape[0] - 2):
        # Find Support levels
        if is_support(df, i):    
            low = df['Low'][i]    
            if is_far_from_level(low, levels, df):      
                levels.append((i, low))  
        # Find Resistence levels
        elif is_resistance(df, i):    
            high = df['High'][i]    
            if is_far_from_level(high, levels, df):      
                levels.append((i, high))
    return levels
````

````python
# get Support & Resistence levels
levels = support_resistence(df_resampled)
````

To plot ***Support & Resistence levels*** we will use `plt.hlines(y, xmin, xmax, colors, linestyles)` method which *plots horizontal lines.*

*   `y` - indexes where to plot the lines.
*   `xmin, xmax` - respective beginning and end of each line
*   `colors` - list of line colors
*   `linestyles` - line style *('solid', 'dashed', 'dashdot', 'dotted')*

````python
# Create a figure and a set of subplots specifying figure size
fig, ax = plt.subplots(figsize=(16, 9))  

# Plot High price
plt.plot(df_resampled.High, color='tab:blue',)
# Plot Low price
plt.plot(df_resampled.Low, color='tab:red')

# Plot Support & Resistence lines
for level in levels:
    plt.hlines(level[1], xmin = df_resampled.index[level[0]], xmax = max(df_resampled.index), colors='green', linestyle='--')

# Set axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')
# Set title
ax.set_title('BTC Cryptocurrency Chart with Support & Resistence')

# Display legend
ax.legend(['High price', 'Low price', 'Support & Resistence'])

# Show plot
plt.show()
````

### Filling area between lines <a id="17"></a>

#### Bollinger Bands

***Bollinger Bands*** are *volatility bands* placed above and below a Moving Average and plotted on price. *Volatility* is based on the *standard deviation*. 

[Why do we need them?](https://www.schwab.com/learn/story/bollinger-bands-what-they-are-and-how-to-use-them)
Bollinger Bands help identify sharp, short-term price movements and potential entry and exit points. Flexible and visually intuitive to many traders, Bollinger Bands can be a helpful technical analysis tool. They can be used to read the trend strength, to time entries during range markets and to find potential market tops.

The bands that often encompass price expand and contract as *volatility expands*, and *decreases* are based on *+2 standard deviations above* the center line ***(top band)*** and *-2 standard deviations below* the center line ***(bottom band).***

Read more [here](https://www.investopedia.com/charts-for-crypto-6500665).

````python
def get_bollinger_bands(price, rate=10):
    """
    Returns Bollinger top and bottom bands
    """
    # Calculate Moving Average
    sma = price.rolling(rate).mean()
    # Calculate volatility     
    std = price.rolling(rate).std()
    
    # Calculate top band
    bollinger_up = sma + std * 2
    # Calculate bottom band
    bollinger_down = sma - std * 2
    
    return bollinger_up, bollinger_down
````

````python
# Get Bollinger top and bottom bands
bollinger_up, bollinger_down = get_bollinger_bands(df_resampled['Close'])
````

Previously, we used *Pandas* `df.plot()` method to plot a line on graph. However, ***matplotlib.pyplot*** also offers this implementatiion as `pyplot.plot(x, y)`. Besides, we can set label and color line by passing parameters as follows `pyplot.plot(x, y, label, color)`.

To identify ***Bollinger Bands area*** we use `ax.fill_between(x, y1, y1)` method where:
*   `x` - the x coordinates of the nodes defining the curves 
*   `y1` - the y coordinates of the nodes defining the first curve
*   `y2` - the y coordinates of the nodes defining the second curve
*   `label` - label of the line
*   `color` - color of the line

````python
# Create a figure and a set of subplots specifying figure size
fig, ax = plt.subplots(figsize=(16, 9))

# Plot Closing price
plt.plot(df_resampled.Close, label='Closing Price', color='tab:blue')
# Plot Bollinger Up band
plt.plot(bollinger_up, label='Bollinger Up', color='darkslateblue')
# Plot Bollinger Down band
plt.plot(bollinger_down, label='Bollinger Down', color='tab:red')

# Set axis labels
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Set title
ax.set_title('BTC Cryptocurrency Chart with Bollinger Bands')

# Fill Bollinger bands area
ax.fill_between(df_resampled.index, bollinger_up, bollinger_down, color='powderblue')

# Display legend
ax.legend()
# Show plot
plt.show()
````
